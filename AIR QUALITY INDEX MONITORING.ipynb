```python
# We have annotations like 1.7#,9.6#,2.3*,2.5x, we want to clean that.
def numeric(row):
    try:
        if np.isnan(row):
            return
        else:
            row =str(row)
            return float(row.replace('x','').replace('#','').replace('*',''))
    except TypeError:
        row =str(row)
        return float(row.replace('x','').replace('#','').replace('*',''))
data_as_frame['WS_HR'] = data_as_frame['WS_HR'].apply(numeric)
```

```python
# Filling nan value with median 

data_as_frame['WS_HR'].fillna(value=data_as_frame['WS_HR'].median(),inplace=True)
```

```python
#For all coloumns filling nan with median 
for col in col_list:
    data_as_frame[col]=data_as_frame[col].apply(numeric)
    data_as_frame[col].fillna(value=data_as_frame[col].median(),inplace=True)
```

```python
# Some are annoted with NR we replace them with 0

data_as_frame['RAINFALL'] = data_as_frame['RAINFALL'].apply(lambda x:0 if x=='NR' else x).apply(numeric)
```

```python
#Significat features as per image significance.png
data = data_as_frame[['CO', 'NO', 'NO2', 'NOx', 'O3', 'PM10', 'PM2.5', 'SO2']]
data.head(10)
```

```python
temp_data = data_as_frame.drop(columns=['CO', 'NO', 'NO2', 'NOx', 'O3', 'PM10', 'PM2.5', 'SO2'],axis=1)
temp_data.head()
```

```python
#AQI is the max of the above significant classes from imgae aqi.png
data['AQI']= data.max(axis=1)
data.head()
```

```python
#AQI is the class to label based table.png

aqi = data['AQI']

aqi[(aqi > 0) & (aqi <= 50)] = 0
aqi[(aqi > 50) & (aqi <= 100)] = 1
aqi[(aqi > 100) & (aqi <= 150)] = 2
aqi[(aqi > 150) & (aqi <= 200)] = 3
aqi[(aqi > 200) & (aqi <= 300)] = 4
aqi[(aqi > 300)] = 5

print(aqi.unique())
```

```python
#Replacce AQI with class and label them 

data['CLASS'] = data['AQI'].astype('int8')
data.drop(columns=['AQI'], axis= 1, inplace=True)
data.head(10)
```

```python
#Training dataset 
```

```python
#Splitting data classes into features as x and Label as y
x = data.drop('CLASS',axis =1)
y = data['CLASS']
```

```python
from keras import models
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from keras.utils import to_categorical
from keras import losses 
from keras import metrics
from keras.models import Sequential
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#Splitting dataset 30%as test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python

train_data = pd.concat([x_train, y_train], axis=1, join='inner')
test_data = pd.concat([x_test, y_test], axis=1, join='inner')
```

```python
#To convert the label into array using One hot encoder where 2=[0,0,1,0,0,0] it 1 =[0,1,0,0,0,0]
Y_train = to_categorical(y=y_train,num_classes=6)
Y_test = to_categorical(y=y_test,num_classes=6)
```

```python
#Defing model 
model =Sequential()
model.add(Dense(8, input_dim =15, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(6, activation='softmax'))
model.compile(
    optimizer='adam',
    loss= 'categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
# train the model
model.fit(
    x_train,
    Y_train,
    epochs = 30,
    batch_size = 4
)
```

```python
#To save model weights and structure
model.save('/content/Model_AirPollution.h5')
```

```python
#Testing dataset
y_pred_rounded = np.argmax(model.predict(x_test), axis=-1)
y_pred_rounded
```

```python
acc = accuracy_score(y_true=y_test, y_pred=y_pred_rounded)*100
print("Accuracy in percentage : %f"%acc)
```

```python
#to Print confusion matrix 
import itertools
def plot_confusion_matrix(cm, classes,title='Confusion matrix'):
    
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    
    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()
```

```python
# Class Label
class_names = ['Good', 'Satisfactory', 'Moderately', 'Poor', 'Very-Poor', 'Severe']
```

```python
print(classification_report(y_true=y_test, y_pred=y_pred_rounded))
```

```python
cm = confusion_matrix(y_pred=y_pred_rounded, y_true=y_test)
plot_confusion_matrix(cm, class_names)
```

```python
import pandas as pd
data_time = pd.read_csv("predicted_data.csv")
```

```python
data_time.head()
```

```python
data_frame = data_time.drop(columns = ["CLASS","predict"])
```

```python
data_frame = data_frame.set_index(data_time.time)
```

```python
data_frame.drop(columns = ["time"], inplace=True)
```

Here we are predicting PM10

```python
from lightgbm import LGBMRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
X = data_frame.drop('PM10', axis=1)
y = data_frame['PM10']
```

```python
#take last week of the dataset for validation
horizon=24*7
X_train, X_test = X.iloc[:-horizon,:], X.iloc[-horizon:,:]
y_train, y_test = y.iloc[:-horizon], y.iloc[-horizon:]
```

```python
#create, train and do inference of the model
model = LGBMRegressor(random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
#calculate MAE
mae = np.round(mean_absolute_error(y_test, predictions), 3)  
mae
```

```python
#plot reality vs prediction for the last week of the dataset
fig = plt.figure(figsize=(16,8))
plt.title(f'Real vs Prediction - MAE {mae}', fontsize=20)
plt.plot(y_test, color='red')
plt.plot(pd.Series(predictions, index=y_test.index), color='green')
plt.xlabel('Time', fontsize=16)
plt.ylabel('PM10', fontsize=16)
plt.legend(labels=['Real', 'Prediction'], fontsize=16)
plt.grid()
plt.show()
```

```python
#create a dataframe with the variable importances of the model
df_importances = pd.DataFrame({
    'feature': model.feature_name_,
    'importance': model.feature_importances_
}).sort_values(by='importance', ascending=False)
```

```python
#plot variable importances of the model
import seaborn as sns
plt.title('Variable Importances', fontsize=16)
sns.barplot(x=df_importances.importance, y=df_importances.feature, orient='h')
plt.show()
```
